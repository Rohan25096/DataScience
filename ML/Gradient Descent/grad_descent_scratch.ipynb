```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

```python
# Now let's apply gradient descent.
# We'll assume slope, m= 78.35 and starting value of intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

# Iteration1

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
learning_rate = 0.1

step_size = learning_rate*loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.plot(X,y_pred2,color='blue',label='b = 20.92')
plt.legend()
plt.show()
```
